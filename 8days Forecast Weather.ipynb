```python
city_name = "singapore" #you can ask for user input instead

#Let's get the city's coordinates (lat and lon)
url = f'https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}'
print(url)

#Let's parse the Json

req=requests.get(url)
data=req.json()

#let's get the name, the longitude and latitudes
name = data['name']
lon = data['coord']['lon']
lat = data['coord']['lat']
print(name,lon,lat)
```

```python
#Let's now use the One Call Api to get the 8 days forecast 
#We'll need to exclude the minutely and hourly
exclude = "minute,hourly"

url2= f'https://api.openweathermap.org/data/2.5/onecall?lat={lat}&lon={lon}&exclude={exclude}&appid={api_key}'
print(url2)

#Let's now parse the Json 
req2= requests.get(url2)
data2=req2.json()

#Let's get the temp for the day, the night and the weather conditions
days=[]
nights=[]
descr = []

#we need to access 'daily'
for i in data2['daily']:
    #We notice that the temperature is in Kelvin, so we need to =-273.15 for every datapoint
    
    #Let's start by days
    #Let's round the decimal to 2
    days.append(round (i['temp']['day'] - 273.15,2))
    
    #Nights
    nights.append (round (i['temp']['night'] -273.15,2))
    
    #Let's now get the weather condition and the description 
    #print (i)
    #'weather'[0] 'main'[0] 'description'
    descr.append(i ['weather'][0]['main'] + ":" + i['weather'][0]['description'])
    
print (days)
print (nights)
print (descr)
```

```python
#Let's now format the output to make it readable 
string = f'[{name} - 8 days forecast]\n'

#Let's now loop for as much days as there are available (8 in this case):
for i in range (len (days)):
    #we want to print the day (day 1,2,3,4... )
    
    if 1==0:
        string += f'\nDay {i +1} (Today)\n'
    elif i == 1:
        string += f'\nDay {i+1} (Tomorrow)\n'
    
    else:
        string += f'\nDay {i+1}\n'
        
    #Now, let's add to that the temperatures
    #°C
    
    string += 'Morning' + str(days [i]) + '°C' + "\n"
    string += 'Night' + str(nights[i]) + '°C'+ "\n"
    string += 'Conditions' + str(descr [i]) + '°C' + '\n'
    
print (string)
```
